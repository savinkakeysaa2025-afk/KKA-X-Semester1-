#5.paramater fungsi print

```python
nama_siswa_1 = "Jielo"
nama_siswa_2 = "Aida"
nama_siswa_3 = "Dani"
nama_siswa_4 = "Nurul"

print(nama_siswa_1, end="/")
print(nama_siswa_2, end="/")
print(nama_siswa_3, end="/")
print(nama_siswa_4)
```

```python
nama_siswa_5 = "Hasbi"
nama_siswa_6 = "Habibi"
nama_siswa_7 = "Rendra"
nama_siswa_8 = "Naufal"

print(nama_siswa_8, nama_siswa_7, nama_siswa_6, nama_siswa_5, sep="*")
```
